```python
import pandas as pd
```

```python
dt_features = pd.read_csv("dt_features.csv").reset_index()
rf_features = pd.read_csv("rf_features.csv").reset_index()
xgb_features = pd.read_csv("xgb_features.csv").reset_index()
```

```python
dt_features.rename(columns={"index": "dt_rank", "features_weights": "dt_weights"}, inplace=True)
rf_features.rename(columns={"index": "rf_rank", "febatures_weights": "rf_weights"}, inplace=True)
xgb_features.rename(columns={"index": "xgb_rank", "features_weights": "xgb_weights"}, inplace=True)
```

```python
all_features = dt_features.merge(rf_features, how= 'left',on='Features')
all_features = all_features.merge(xgb_features, how= 'left',on='Features')
```

```python
all_features['final_rank'] = all_features['dt_rank'] + all_features['rf_rank'] + all_features['xgb_rank']
all_features.sort_values(by=['final_rank'], ascending=True, inplace=True)
all_features.reset_index().drop(columns=['index'])
```

```python
ranked_features = all_features[['Features', 'final_rank']].reset_index().drop(columns=['index'])
ranked_features
```
